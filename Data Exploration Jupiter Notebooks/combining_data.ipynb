```python
import pandas as pd
from pathlib import Path
```

```python
video_game_ratings_unclean_path = Path('..//Uncleaned Data//partially_cleaned_video_games.csv')
video_game_sales_path = Path('..//Uncleaned Data//Video game sales kaggle//video_games_sales.csv')
ratings_df = pd.read_csv(video_game_ratings_unclean_path)
sales_df = pd.read_csv(video_game_sales_path)
```

```python
ratings_df.head()
```

```python
sales_df.head()
```

```python
#drop null values from rating column
ratings_df = ratings_df.dropna(subset=['rating'])

ratings_df.head(10)
```

```python
#merge the two dataframes together based on name
merged_df = pd.merge(ratings_df, sales_df, left_on='name', right_on='name', how='inner')
merged_df.head(20)
```

```python
merged_df.info()
```

```python
merged_df.to_csv('../Cleaned Data/ratings_and_sales.csv')
```
