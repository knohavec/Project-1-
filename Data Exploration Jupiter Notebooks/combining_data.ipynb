```python
import pandas as pd
from pathlib import Path
```

```python
video_game_ratings_clean_path = Path('..//Cleaned Data//video_games_ratings.csv')
video_game_sales_path = Path('..//Cleaned Data//video_games_sales.csv')
ratings_df = pd.read_csv(video_game_ratings_clean_path)
sales_df = pd.read_csv(video_game_sales_path)
```

```python
ratings_df.head()
```

```python
sales_df.head()
```

```python
#drop null values from rating column
ratings_df = ratings_df.dropna(subset=['rating'])

ratings_df.head(10)
```

```python
#merge the two dataframes together based on name
merged_df = pd.merge(ratings_df, sales_df, left_on='name', right_on='name', how='inner')
merged_df.head(20)
```

```python
merged_df.info()
```

```python
merged_df.to_csv('../Cleaned Data/ratings_and_sales.csv')
```

```python
ratings_and_sales = pd.read_csv('../Cleaned Data/ratings_and_sales.csv')
ratings_and_sales.head()
```

```python
#split the games into 10 groups based on rating (0-10, 11-20, ... 91-100)
ratings_and_sales['rating_bins'] = pd.cut(ratings_and_sales['rating'], bins=10, labels=False) * 10
ratings_and_sales.head()
```

```python
# Calculate the average global sales for each rating bin
avg_sales_per_bin = ratings_and_sales.groupby('rating_bins')['global_sales'].mean()
```

```python
#plot the rating bins vs the average global sales of games in that bin
figure2 = avg_sales_per_bin.plot(kind='bar', figsize=(10, 6))
figure2.set_xlabel("Ratings of games as a %")
figure2.set_ylabel('Average global sales for the rating group (millions of copies sold)')
figure2.set_title('Ratings vs Sales')
figure2.set_xticklabels([f'{x}-{x+9}' for x in range(0, 100, 10)], rotation=45)
```
